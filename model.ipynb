```
# import packages used
import pandas as pd

### 讀入series
# df = pd.read_csv('stocker/price.csv', index_col='date', parse_dates=['date'])
google = pd.read_csv('stocker/google.csv')
amazon = pd.read_csv('stocker/amazon.csv')
fb = pd.read_csv('stocker/facebook.csv')
# print (df)
from stocker.stocker import Stocker
# price = df.squeeze()
# price.head()
google = Stocker(google, trainingYears=8)
amazon = Stocker(amazon, trainingYears=8)
fb = Stocker(fb, trainingYears=5)
```

```
googleModel, googleModelData = google.create_prophet_model(days=90)
amazonModel, amazonModelData = amazon.create_prophet_model(days=90)
fbModel, fbModelData = fb.create_prophet_model(days=90)
```

```
googleLabel = pd.read_csv('./stocker/googleLabel.csv', parse_dates=['Date'])
amazLabel = pd.read_csv('./stocker/amazLabel.csv', parse_dates=['Date'])
fbLabel = pd.read_csv('./stocker/fbLabel.csv', parse_dates=['Date'])
# googleLabel = googleLabel[googleLabel['Date'] <= googleLabel['Date'][0] + pd.DateOffset(days=5)]
googleLabel['Price'] = googleLabel['Close']
amazLabel['Price'] = amazLabel['Close']
fbLabel['Price'] = fbLabel['Close']

googleLabel['diff'] = googleLabel['Close'].diff()
googleLabel['direction'] = (googleLabel['diff'] > 0) *1
amazLabel['diff'] = amazLabel['Close'].diff()
amazLabel['direction'] = (amazLabel['diff'] > 0) *1
fbLabel['diff'] = fbLabel['Close'].diff()
fbLabel['direction'] = (fbLabel['diff'] > 0) *1


googleLabel = googleLabel.dropna()
amazLabel = amazLabel.dropna()
fbLabel = fbLabel.dropna()
# print(googleLabel['Date'])
```

```
google.changepoint_prior_analysis()
amazon.changepoint_prior_analysis()
fb.changepoint_prior_analysis()
```

```
google.changepoint_prior_scale = 0.2
amazon.changepoint_prior_scale = 0.2
fb.changepoint_prior_scale = 0.2

google.predict_future(googleLabel, days=90)
amazon.predict_future(amazLabel, days=90)
fb.predict_future(fbLabel, days=90)
```
